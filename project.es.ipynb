#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Código
import pandas as pd
grades = [8.5,9.2,7.8,8.9,10]
serie_grades = pd.Series(grades)
promedio = serie_grades.mean()
print(promedio)
```

- Encuentra la calificación más alta y la más baja.

```python
# Código
import pandas as pd
maximo = serie_grades.max()
minimo = serie_grades.min()
print("Calificación máxima=", maximo)
print("Calificación mínima=", minimo)
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
# Trazar el histograma de 
import numpy as np
ds = pd.read_csv('assets/real_estate.csv', sep=';')
arroyomolinos = ds['level5'] == "Arroyomolinos (Madrid)"
ds_arroyo = ds[arroyomolinos]
precio_arroyo = ds_arroyo['price'].dropna()
plt.Figure(figsize=(10,5))
plt.hist(precio_arroyo,bins=16)
plt.title('Histograma de Precios Arroyomolinos (Madrid)')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
```

El histograma obtenido indica que el grueso de los precios de propiedades en Arroyomolinos oscila entre los 150000 y 400000 euros, con una concentración de propiedades de dentro del rango de 350000 y 375000. Hay un segundo rango de precios, que concentra una frecuencia cercana a las 50 propiedades, cuyos precios oscilan entre los 250000 y 300000 Euros. Se puede apreciar también que hay un pequeño grupo de menos de cinco propiedades con precios inferiores a los 50000 Euros, y sería interesante revisar estos casos, que pueden corresponder a propiedades en alquiler.
